```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101, size = (50,5))
df = pd.DataFrame(scores, 
                  columns = ['國文', '英文', '數學', '地理', '歷史'], 
                  index = range(1,51)
                  )

df.index.name = '學號'
df.columns.name = '科目'
df
```

```python
sum_value = df.sum(axis = 1)  # 沿著行的方向加總
mean_value = df.mean(axis = 1)  # 沿著行的方向平均
df['總分'] = sum_value
df['平均'] = mean_value
df1 = df.sort_values(by='總分',ascending=False) # 依照總分排序
df1
```

```python
rand_value = df1['總分'].rank(ascending=False,method='min') # 依照總分排名, 同分者名次相同
df1['排名'] = rand_value
df1['排名'] = df1['排名'].astype(int) # 將排名轉為整數
df1
```

```python
def max_score(row) -> int:
    return row.max()

df1['最高分'] = df1[['國文', '英文', '數學', '地理', '歷史']].apply(max_score, axis=1)
df1
```

```python
def min_score(row) -> int:
    less_than_59 = row < 60
    return row[less_than_59].count()

df1['不及格科目數量'] = df1[['國文', '英文', '數學', '地理', '歷史']].apply(min_score, axis=1)
df1
```

```python
def max_score(row):
    less_60_count = len(row[row<60])
    return pd.Series([row.min(), row.max(), row.std(),less_60_count], index=['最低分', '最高分', '標準差', '不及格科目數'])

df2 = df1[['國文', '英文', '數學', '地理', '歷史']].apply(max_score, axis=1)
df3 = pd.concat([df1, df2], axis=1)
df3
```
