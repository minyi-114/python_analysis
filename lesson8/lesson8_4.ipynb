```python
df2 = df1.rename(columns=df1.iloc[0].to_dict())
df2.head()
```

```python
df3 = df2.drop(index=0)
df3.head()
```

```python
df3.info()
```

```python
import numpy as np

def str_to_int(value:str) -> int: #將字串轉換為整數
    try:
        value_int =int(value)
    except:
        value_int = 0
    return value_int

def str_to_float(value:str) -> float: #將字串轉換為浮點數
    try:
        value_float = float(value)
    except:
        value_float = 0
    return value_float

df3['年底人口數'] = df3['年底人口數'].map(str_to_int) #將 '年底人口數' 欄位的資料轉換為整數
df3['人口密度'] = df3['人口密度'].map(str_to_int)  #將 '人口密度' 欄位的資料轉換為整數
df3['土地面積'] = df3['土地面積'].map(str_to_float)  #將 '土地面積' 欄位的資料轉換為浮點數
df3.info()
```

```python
df3.tail()
```

```python
def split_city(city:str) -> str: #取得縣市名稱
    if len(city) > 4:
        return city[:3]
    else:
        return city 
    
def split_area(area:str) -> str: #取得區域名稱
    if len(area) > 4:
        return area[3:]
    else:
        return '全區'
    
df3['縣市'] = df3['區域別'].map(split_city) #取得縣市名稱
df3['鄉鎮市區'] = df3['區域別'].map(split_area) #取得區域名稱
df4 = df3.reindex(columns=['縣市', '鄉鎮市區', '年底人口數', '土地面積', '人口密度']) #重新排序欄位
df4.tail()
```

```python
df4.info()
```

```python
def population(value:int) -> str: #取得人口數的描述
    if value > 367653:
        return '人多'
    elif value > 183826:
        return '適中'
    else:
        return '人少'
    
df4['人口分佈'] = df4['年底人口數'].map(population) #取得人口數的描述
df4.loc[df4['縣市']=='臺中市'].head(10)
```

```python
import seaborn as sns

# ... (Your existing code from lesson8_4.ipynb) ...

# Select relevant columns
df_plot = df4[['縣市', '年底人口數']]

# Create the histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=df_plot, x='年底人口數', hue='縣市', multiple="stack", kde=False)
plt.xlabel('年底人口數')
plt.ylabel('次數')
plt.title('各縣市人口數量分布圖')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


#Alternative: Boxplot for better comparison of medians and ranges
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_plot, x='縣市', y='年底人口數')
plt.xlabel('縣市')
plt.ylabel('年底人口數')
plt.title('各縣市人口數量盒子圖')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
