```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mlp
fontManager.addfont('ChineseFont.ttf')
mlp.rc('font', family='ChineseFont')
```

```python
figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1,1,1)
axes.set_xticks([1, 3, 5, 7, 9])
axes.set_yticks([-1, -0.5, 0, 0.5, 1])
plt.show()
```

```python
figure = plt.figure(figsize=(10,6))
axes = figure.add_subplot(1,1,1)
axes.plot([11, 3, 5, 7, 9],marker='o',color='red', linestyle=':')  # ':' dotted, '--' dashed
plt.show()
```

```python
figure = plt.figure(figsize=(10,6))
ax1 = figure.add_subplot(2,1,1)
ax1.plot([11, 3, 5, 7, 9],marker='x',color='green', linestyle='')
ax2 = figure.add_subplot(2,1,2)
ax2.bar(['a', 'b', 'c', 'd', 'e'],[1, 2, 4, 3, 5])
plt.show()
```

```python
student1 = {'':78, '':80, '':92, '':75, '':85}
student2 = {'':90, '':75, '':98, '':65, '':75}
```

```python
subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,8))
axes = figure.add_subplot(1,1,1)
axes.set_title("")
axes.set_xlabel("")
axes.set_ylabel("")
axes.plot(subjects,score1,'ro-.',label="")
axes.plot(subjects,score2,'gx--',label="")
# axes.bar(subjects,score1)
axes.legend()  # 
axes.grid(axis='y')  # 
plt.show()
```
