```python
student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
```

```python
subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,8))
axes = figure.add_subplot(1,1,1)
axes.set_title("大明與小明分數比較圖")
axes.set_xlabel("科目")
axes.set_ylabel("分數")

# 使用annotate()標示數據
axes.annotate(f"{student2['歷史']}分退步了", 
              xy=(subjects[3], score2[3]), xytext=(subjects[2], score2[3]+4),
              arrowprops=dict( #arrowstyle="->",
                              facecolor='blue',shrink=0.05, 
                              connectionstyle="arc3,rad=0.3")) #arrowprops 讓標註有箭頭

# 使用text()標示數據
for i, score in enumerate(score1):
    axes.text(subjects[i], score - 1, str(score), ha='center', va='top') #ha水平對齊，va垂直對齊

axes.plot(subjects,score1,'ro-.',label="大明")
axes.plot(subjects,score2,'gs--',label="小明")
# axes.bar(subjects,score1)
axes.legend()  # 顯示圖例
axes.grid(axis='y')  # 顯示格線
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import fontManager
import matplotlib as mlp
fontManager.addfont('ChineseFont.ttf')
mlp.rc('font', family='ChineseFont')

student1 = {'國文':78, '英文':80, '數學':92, '歷史':75, '探索':85}
student2 = {'國文':90, '英文':75, '數學':98, '歷史':65, '探索':75}
subjects = list(student1.keys())
score1 = list(student1.values())
score2 = list(student2.values())

figure = plt.figure(figsize=(10,8))
axes = figure.add_subplot(1,1,1)
axes.set_title("小明和大明分數比較表")
axes.set_xlabel("科目")
axes.set_ylabel("分數")
axes.plot(subjects,score1,'ro-.',label="小明")
axes.plot(subjects,score2,'gx--',label="大明")
axes.legend()
axes.grid(axis='y')

# 找出數學在subjects列表中的索引
math_index = subjects.index('數學')

for idx, (s, v1, v2) in enumerate(zip(subjects, score1, score2)):
    if s == '數學' and v1 == 92:
        # 對於數學成績，使用特殊的箭頭樣式
        # 使用arc3連接樣式並調整弧度來模擬90度轉向
        axes.annotate(f'{v1}退步了', 
                      xy=(math_index, v1), 
                      xytext=(math_index+0.5, v1),  # 將文字向右上方移動
                      arrowprops=dict(
                          facecolor='red',
                          shrink=0.05,
                          connectionstyle='arc3'  # 負值使弧度向右彎曲
                      ))
        axes.annotate(f'{v2}', xy=(s, v2+0.5))
    else:
        axes.annotate(f'{v1}', xy=(s, v1+0.5))
        axes.annotate(f'{v2}', xy=(s, v2+0.5))
plt.show()
```
